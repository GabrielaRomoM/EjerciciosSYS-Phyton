Interpretación Gráfica

```python
import numpy as np
```

Primer Caso

```python
# Definir las señales x(t) y h(t)
t = np.linspace(0, 1, 100)  # Crear un vector de tiempo de 0 a 2 segundos
x_t = np.zeros_like(t)      # Inicializar x(t) como un vector de ceros del mismo tamaño que t
x_t[(t >= 0) & (t <= 1)] = 1  # Definir la señal x(t) como rect(t-0.5)
h_t = np.exp(-t)            # Definir la señal h(t) como e^-t

# Realizar la convolución
convolution_result = np.convolve(x_t, h_t, mode='full') / len(t)  # Normalizar el resultado

# Graficar las señales y el resultado de la convolución
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
plt.subplot(311)
plt.plot(t, x_t)
plt.title('Señal x(t)')

plt.subplot(312)
plt.plot(t, h_t)
plt.title('Señal h(t)')

plt.subplot(313)
t_conv = np.linspace(0, 2, len(convolution_result))  # Crear un vector de tiempo para el resultado de la convolución
plt.plot(t_conv, convolution_result)
plt.title('Resultado de la convolución')

plt.tight_layout()
plt.show()
```
